```python
from flask import Flask, jsonify
import random

app = Flask(__name__)

@app.route('/risk')
def risk_score():
    score = round(random.uniform(0.1, 0.9),2)
    return jsonify({"risk_score": score})

# app.run(debug=True)
print("Flask API ready at /risk")
```

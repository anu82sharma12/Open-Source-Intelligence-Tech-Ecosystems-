```python
import random
langs = ["Hindi","Tamil","Bengali","Kannada","Urdu","Gujarati","Malayalam","Odia","Assamese","Marathi","Telugu","Manipuri"]
tokens = [f"{lang}_word_{i}" for lang in langs for i in range(10)]
print("Sample Tokens:", random.sample(tokens,5))
print(f"Total Tokens: {len(tokens)}")
```
